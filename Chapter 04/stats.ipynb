```python
fn=0
plt.title(iris.feature_names[fn])
plt.xlabel('Intervals')
plt.ylabel('Frequency')
f = plt.hist(iris.data[:, fn], bins = 10)
max_freq = np.max(f[0])
interval_num = np.argmax(f[0])
h=f[1][1]-f[1][0]
l= f[1][interval_num]
mode = l + h/2
plt.vlines(x=np.mean(iris.data[:, fn]),
           ymin=0, ymax=max_freq, label='mean', linestyles='solid')
plt.vlines(x=np.median(iris.data[:, fn]), 
           ymin=0, ymax=max_freq, label='median', linestyles='dashed')
plt.vlines(x=mode, 
           ymin=0, ymax=max_freq, label='mode-1', linestyles='dotted')
plt.vlines(x=5, 
           ymin=0, ymax=max_freq, label='mode-2', linestyles='dotted')

plt.legend()
#plt.savefig('iris-'+str(fn)+'.png')
```

```python
y1 = np.random.binomial(n=20, p=0.1, size=100)
y2 = np.random.binomial(n=20, p=0.3, size=100)
y3 = np.random.binomial(n=20, p=0.5, size=100)
y4 = np.random.binomial(n=20, p=0.7, size=100)
y5 = np.random.binomial(n=20, p=0.9, size=100)
```

```python
y = [y1, y2, y3, y4,y5]
```

```python
fig, axes = plt.subplots(ncols=5, nrows=1)

for i, ax in zip(range(10), axes.flat):
    sns.displot(y[i], binwidth=1, ax=ax)
    plt.savefig('binomial_{}.png'.format(i), dpi=400)
    plt.show()
```

```python
fig, axes = plt.subplots(ncols=4, nrows=1)

lambdas = [1, 4, 10, 20]
for i, ax in zip(range(10), axes.flat):
    y = np.random.poisson(lam=lambdas[i], size=100)
    sns.distplot(y, hist=False, ax=ax)    
    plt.show()
```
