```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def tanh(z):
    ez = np.exp(z)
    enz = np.exp(-z)
    return (ez - enz)/ (ez + enz)

def sigmoid(z):
    ez = 1+np.exp(-z)
    return 1/ez

# Calculate plot points
z = np.arange(-3., 3., 0.01)
a = tanh(z)
dz = 1 - a**2

b = sigmoid(z)
dz2 = b*(1-b)
```

```python
plt.figure(figsize=(15,5))
# Setup centered axes
ax = plt.subplot(1, 2, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Create and show plot
ax.plot(z,a, color="black", linewidth=3, label="$y=tanh(x)$")
ax.plot(z,dz, color="#9621E2", linewidth=3, label="$tanh'(x)$")
ax.legend(loc="upper left", frameon=False)


# Setup centered axes
ax1= plt.subplot(1, 2, 2)
ax1.spines['left'].set_position('center')
ax1.spines['bottom'].set_position('center')
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

# Create and show plot
ax1.plot(z,b, color="black", linewidth=3, label="$y=\sigma(x)$")
ax1.plot(z,dz2, color="#9621E2", linewidth=3, label="$\sigma'(x)$")
ax1.legend(loc="upper left", frameon=False)
#plt.show()
plt.savefig('derivative.png', dpi=300)
```
