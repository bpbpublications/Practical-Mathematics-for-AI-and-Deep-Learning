```python
import statsmodels.api as sm
```

```python
series = pd.read_csv('AirPassengers.csv', header=0)
df = pd.DataFrame(series)
```

```python
df.columns
```

```python
df = df.reset_index(drop=True).set_index('Month')
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df.index.asfreq('M')
```

```python
res = sm.tsa.seasonal_decompose(df, freq=12)
```

```python
resplot = res.plot()
```
